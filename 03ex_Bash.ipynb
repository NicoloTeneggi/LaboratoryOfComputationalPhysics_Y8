### Ex 1

1\.a Make a new directory called `students` in your home. Download a csv file with the list of students of this lab from [here](https://www.dropbox.com/s/867rtx3az6e9gm8/LCP_22-23_students.csv) (use the `wget` command) and copy that to `students`. First check whether the file is already there

1\.b Make two new files, one containing the students belonging to PoD, the other to Physics.

1\.c For each letter of the alphabet, count the number of students whose surname starts with that letter. 

1\.d Find out which is the letter with most counts.

1\.e Assume an obvious numbering of the students in the file (first line is 1, second line is 2, etc.), group students "modulo 18", i.e. 1,19,37,.. 2,20,38,.. etc. and put each group in a separate file  

```python
mkdir students

#check if the file is already there, 

if [ ! -f './students.csv' ]
then
    touch students.csv
fi

#download the file and copy it to students.csv
wget -O students.csv https://www.dropbox.com/scl/fi/bxv17nrbrl83vw6qrkiu9/LCP_22-23_students.csv?rlkey=47fakvatrtif3q3qw4q97p5b7&e=1

#create the new files
touch students_PoD.csv
touch students_Phys.csv

#copy the correct students
grep 'PoD' students.csv > students_PoD.csv
grep 'Physics' students.csv > students_Phys.csv

#count!
for x in {A..Z}; do echo $x; grep -c  ^$x students.csv; done

#inizialize 'maximum' variables
max=0
letter=0
#compare each count to the maximum
for x in {A..Z}; do echo $x; grep -c  ^$x students.csv; if [ $(grep -c  ^$x students.csv) -gt $max ]; then max=$(grep -c  ^$x students.csv); letter=$x; fi; done
echo $max
echo $letter

#check total number of students
total=$(grep -c -v 'E-mail' students.csv)

#create files and append lines ->   THE FIRST USEFUL LINE IS LINE 2

for x in {1..18}; do touch group$x.csv; counter=$(($x + 1)); while [ $counter -lt $total ]; do cut -f $counter -d $'\n' students.csv >> group$x.csv; let counter+=18; done; done
```
